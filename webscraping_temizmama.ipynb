```python
l=[]

for item in all:

    d={}
    d["isim"]=item.find("span",{"itemprop":"name"}).find_all("b")[0].text
    d["tanim"]=item.find("span",{"itemprop":"name"}).find_all("b")[1].text

    try:
        paket1tip=item.find("td",{"class":"font12"}).text.replace(" ","")
        paket1tip=paket1tip.split('\r\n\r\n')[0]
        paket1tip=paket1tip.replace("\r\n","")
        paket1tip=paket1tip.split('\n\n')[0]
        d["paket1tip"]=paket1tip
    except:
        d["paket1tip"]=None

    try:
        d["uretim1"]=item.find("td",{"class":"font12"}).find("div",{"class":"r-urun-m-uretim visible-xs"}).text.replace("\r","").replace("\n","").replace("\t","").replace(" ","").replace("günönceüretildi","")
    except:
        d["uretim1"]=None
    
    try:
        d["paket1_tekfiyat"]=item.find("td",{"class":"font12"}).find("div",{"class":"urun-paket"}).text.replace("1 Adet = ","").replace("(","").replace(")","")
    except:
        d["paket1_tekfiyat"]=None
    
    try:
        d["paket1_toplamfiyat"]=item.find("td",{"itemprop":"price"}).text
    except:
        d["paket1_toplamfiyat"]=None
        
    try:
        paket2tip=item.find("td",{"class":"font15"}).text.replace(" ","")
        paket2tip=paket2tip.partition('\r\n\r\n')[0].replace("\r\n", "")
        paket2tip=paket2tip.partition('\n\n')[0]
        d["paket2tip"]=paket2tip
    except:
        d["paket2tip"]=None
        
    try:   
        d["uretim2"]=item.find("td",{"class":"font15"}).find("div",{"class":"r-urun-m-uretim visible-xs"}).text.replace("\r","").replace("\n","").replace("\t","").replace(" ","").replace("günönceüretildi","")
    except:
        d["uretim2"]=None
        
    try:
        d["paket2_toplamfiyat"]=item.find_all("td",{"itemprop":"price"})[1].text
    except:
        d["paket2_toplamfiyat"]=None
    
    l.append(d)
    
import pandas
df=pandas.DataFrame(l)
df
```
